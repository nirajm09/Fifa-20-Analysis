```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\DELL\Desktop\Projects\players_20.csv')
```

```python
main = data.drop(columns = ['sofifa_id','long_name','player_url','dob','joined'])
main.head()
```

```python
top_twenty = main.iloc[:20,:8]
top_twenty
```

```python
cols = list(main.columns)
main = main[cols[0:2]+cols[4:7]+cols[2:4]+cols[8:]]
main.head()
```

```python
min_wages = main.sort_values(['wage_eur'],ascending=True)
min_wages = min_wages[(min_wages['wage_eur']>0)]
min_wages.head()
```

```python
braz = main[(main['nationality'] == 'Brazil')]
brazil = braz.iloc[:20]
brazil.head()
```

```python
braznarg = main[(main['nationality'] == 'Brazil') | (main['nationality'] == 'Argentina')]
braznarg  = braznarg.iloc[:20]
braznarg.head()
```

```python
x = braznarg['short_name']
y = braznarg['value_eur']
plt.figure(figsize=(15,8))
plt.xlabel('Player')
plt.ylabel('Market Value')
plt.ylim(1500000,105500000)
plt.bar(x,y)
plt.xticks(rotation=60)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')

bar = main[(main['club'] == 'FC Barcelona')]
barcelona = bar.iloc[:]
barcelona = barcelona.drop(barcelona.ix[:,'international_reputation':'rb'].head(0).columns, axis=1)
barcelona.head()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='short_name',y='wage_eur',data=barcelona).set_title("WAGES of Players at FC Barcelona")
plt.xlabel('Players')
plt.ylabel('Wages in Euros')
plt.xticks(rotation=45)
```

```python
best_young = main[(main['age']<25) & (main['overall']>80)]
y=best_young.sort_values('potential',ascending=False).iloc[:10]
plt.figure(figsize=(10,20))
sns.barplot(x = 'short_name', y = 'potential',data=y)
plt.ylim(85,100)
plt.xlabel('Players')
plt.ylabel('Potential')
plt.xticks(rotation=50)
```
